# Matplotlib example workflow 

```python
# 0. import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data 
x=[1,2,3,4]
y=[11,22,33,44]

# 2. Setup plot
fig, ax=plt.subplots(figsize=(5,5))
# 3 plot the data 
ax.plot(x,y)

# 4. customize plot
ax.set(title= "Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save and show 
fig.savefig("images/matplot1.png")
```

## Two options for sub plot

```python
# subplot option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2, figsize=(5,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nutbutter.keys(),nutbutter.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot Option 2
fig,ax=plt.subplots(nrows=2,ncols=2, figsize=(10,5))
# plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nutbutter.keys(),nutbutter.values());
ax[1,1].hist(np.random.randn(1000));
```

### which one should you use? (pyplot vs matplotlib oo method)

```python
over_50=heart_disease[heart_disease["age"]>50]
```

```python
len(over_50)
```

```python
# pyplot method
over_50.plot(kind="scatter", x="age",y="chol",c="target");
```

```python
# Our Method to plot the above
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            y="chol",
            x="age",
            c="target",
            ax=ax);
```

```python
fig,ax=plt.subplots(figsize=(10,6))
# plot the data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],
                  c=over_50['target']);
# customize the plot 
ax.set(title="Heart Disease and cholesterol levels",
       xlabel="age",
       ylabel="cholesterol");
# Add a legend
ax.legend(*scatter.legend_elements(),title="Target");
# Add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
over_50.head()
```

```python
# subplot of chol,age,thalah

fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],
                   c=over_50["target"])
# customize ax0
ax0.set(title="Heart Disease with age and cholesterol",
      ylabel="Cholesterol Level");
# adding legend
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

# Add data to ax1
scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],
                   c=over_50["target"]);
# Customize ax1
ax1.set(title="Heart Diesease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate ")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");
# adding title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16, fontweight="bold");
```

```python
# subplot of chol,age,thalah
# customizing the y and x axis
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter")
# customize ax0
ax0.set(title="Heart Disease with age and cholesterol",
      ylabel="Cholesterol Level");
# change the x axis limit
ax0.set_xlim([50,80])
# adding legend
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

# Add data to ax1
scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],
                   c=over_50["target"],
                   cmap="winter");
# Customize ax1
ax1.set(title="Heart Diesease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate ")
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");
# adding title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16, fontweight="bold");
```
